# 第六章 练习解答（Chapter 6 Exercise Solutions）


## 练习 6.1：增加上下文长度（Exercise 6.1: Increasing the Context Length）


我们可以通过将最大长度设置为 1024，将输入填充到模型支持的最大 token 数量：

```python
max_length = 1024

train_dataset = SpamDataset(base_path / "train.csv", max_length=max_length, tokenizer=tokenizer)
val_dataset = SpamDataset(base_path / "validation.csv", max_length=max_length, tokenizer=tokenizer)
test_dataset = SpamDataset(base_path / "test.csv", max_length=max_length, tokenizer=tokenizer)
```

或者，我们也可以通过以下方式定义 `max_length`：

```python
max_length = model.pos_emb.weight.shape[0]
```

或者

```python
max_length = BASE_CONFIG["context_length"]
```

为了方便，你可以

```bash
python additional-experiments.py --context_length "model_context_length"
```

使用 [../02_bonus_additional-experiments](../02_bonus_additional-experiments) 文件夹中的代码运行此实验，  
但结果会导致测试准确率显著下降，为 78.33%（相比主章节中的 95.67%）。


## 练习 6.2：对整个模型进行微调（Exercise 6.2: Finetuning the Whole Model）


我们可以不只微调最后一个 Transformer 块，而是微调整个模型，方法是从代码中移除以下几行：


```python
for param in model.parameters():
    param.requires_grad = False
```

为了方便，你可以通过以下方式运行此实验：


```bash
python additional-experiments.py --trainable_layers all
```

使用 [../02_bonus_additional-experiments](../02_bonus_additional-experiments) 文件夹中的代码运行此实验，  
测试准确率提升约 1%，达到 96.67%（相比主章节中的 95.67%）。


## 练习 6.3：微调第一个 token 与最后一个 token（Exercise 6.3: Finetuning the First versus Last Token）


我们可以不微调最后一个输出 token，而是微调第一个输出 token，方法是修改以下内容：

```python
model(input_batch)[:, -1, :]
```

to

```python
model(input_batch)[:, 0, :]
```

在代码中的所有相关位置都进行相应修改。

为了方便，你可以通过以下方式运行此实验：


```
python additional-experiments.py --trainable_token first
```

使用 [../02_bonus_additional-experiments](../02_bonus_additional-experiments) 文件夹中的代码运行此实验，  
测试准确率显著下降，为 75.00%（相比主章节中的 95.67%）。
